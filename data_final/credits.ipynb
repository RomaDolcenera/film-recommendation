```python
import pandas as pd
import ast
```

```python
# Cargo el dataset de credits
dfc = pd.read_csv('../credits.csv')
```

```python
dfc.head(3)
```

```python
# Creo una columna a que llamo actores
dfc['actores'] = ''
```

```python
# Creo una funcion para extraer todos los actores de la columna cast y grabarlo en una nueva columna llamada 'Actors'
for index in dfc['cast'].index:
    lista_diccionarios = ast.literal_eval(dfc['cast'][index])
    actors = []
    for dict in lista_diccionarios:
        for key, value in dict.items():
            if key == 'name':
                actors.append(value)
    dfc.at[index, 'actores'] = actors
```

```python
#Verifico la carga correcta de actores en la nueva columna
dfc['actores']
```

```python
#Verifico el funcionamiento correcto de la busqueda de actores con el nombre Tom Cruise mediante la funcion lambda
result_actor = dfc.loc[dfc['actores'].apply(lambda x: 'Tom Cruise' in x)]
```

```python
result_actor
```

```python
# Creo una funcion para extraer todos los directores de la columna crew y grabarlo en una nueva columna llamada 'directors'
for index in dfc['crew'].index:
    lista_diccionarios = ast.literal_eval(dfc['crew'][index])
    found_condition = False
    directors = []
    for dict in lista_diccionarios:
        for key, value in dict.items():
            if found_condition:
                directors.append(value)
                found_condition = False
                break
            if key == 'job' and value == 'Director':
                found_condition = True
    dfc.at[index, 'directores'] = directors
```

```python
#Verifico la carga correcta de directores en la nueva columna
dfc['directores']
```

```python
#Chequeo el funcionamiento correcto de la busqueda de directores con el nombre Oliver Stone mediante la funcion lambda.
result_director = dfc.loc[dfc['directores'].apply(lambda x: 'Oliver Stone' in x)]
```

```python
dfc.drop(columns=['cast', 'crew'], inplace=True)
```

```python
def remover_corchetes(lista):
    if isinstance(lista, list):
        return ', '.join(lista)  
    return lista
```

```python
dfc['actores'].apply(remover_corchetes)
```

```python
dfc['directores'].apply(remover_corchetes)
```

```python
# Guardo el dataset en un nuevo archivo csv.
dfc.to_csv('../credits_clean.csv', index=False)
```
